```python
import os

# Path to the folder containing label files
label_path = r''  # Adjust this to your label directory

# Function to process label files
def update_labels_for_sniper_and_sword(label_path):
    for split in ['train', 'val']:  # Assuming 'train' and 'val' directories
        split_path = os.path.join(label_path, split)
        for label_file in os.listdir(split_path):
            if label_file.startswith("Sniper") or label_file.startswith("Sword"):  # Check for file name
                label_file_path = os.path.join(split_path, label_file)
                with open(label_file_path, 'r') as file:
                    lines = file.readlines()

                # Update all class IDs to 0
                updated_lines = []
                for line in lines:
                    parts = line.split()
                    parts[0] = '0'  # Change the class ID to 0
                    updated_lines.append(' '.join(parts) + '\n')

                # Write the updated labels back to the file
                with open(label_file_path, 'w') as file:
                    file.writelines(updated_lines)

                print(f"Updated {label_file_path}")

# Run the function
update_labels_for_sniper_and_sword(label_path)
```

```python
import os

# Path to your labels directory
labels_path = "/content/drive/MyDrive/dataset/labels/val"

for file_name in os.listdir(labels_path):
    if file_name.endswith(".txt"):
        file_path = os.path.join(labels_path, file_name)
        with open(file_path, "r") as file:
            lines = file.readlines()

        # Update classes
        updated_lines = []
        for line in lines:
            parts = line.strip().split()
            class_id = int(parts[0])
            if class_id in [1,2,3,4,5,6,7,8,9]:  # Change 1 and 2 to 0
                parts[0] = "0"
            updated_lines.append(" ".join(parts))

        # Write updated labels back to file
        with open(file_path, "w") as file:
            file.write("\n".join(updated_lines))

```
